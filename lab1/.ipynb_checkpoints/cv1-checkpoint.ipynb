```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(im[:,:,0])
plt.subplot(222)
plt.imshow(im[:,:,1])
plt.subplot(223)
plt.imshow(im[:,:,2])
plt.subplot(224)
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('.\\6_26.jpg')
fig = plt.figure(figsize=(15,15))

plt.imshow(im[350:600,900:1150])
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
im[250:500, 250:500] = im[350:600, 900:1150] 
im[350:600, 900:1150] = im[0:250, 0:250]
plt.imshow(im)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
temp = im[:,:,0]
im[:,:,0] = im[:,:,2]
im[:,:,2] = im[:,:,1]
im[:,:,1] = temp
plt.imshow(im)
plt.show()
```
